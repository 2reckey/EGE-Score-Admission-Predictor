```python
import pandas as pd
import numpy as np
```

```python
parts = []
for i in range(10):
    part = pd.read_csv(f'Data/full_data_{i}.csv')
    parts.append(part)

df = pd.concat(parts, ignore_index=True)
df
```

```python
df['All_app']/=df['Quota']
df['Orig_all']/=df['Quota']
df['BVI']/=df['Quota']
df['Orig_above']/=df['Quota']
```

```python
import seaborn as sns

sns.histplot(df[~df['Is_pass']]["Avg_score"])
```

```python
df['Combined'] = df.apply(lambda row: f"{row['Year']} - {row['University']} - {row['Faculty']} - {row['Program']}", axis=1)
unique_values = df['Combined'].unique()
df
```

```python
from sklearn.model_selection import train_test_split

train_indices, test_indices = train_test_split(unique_values, test_size=0.2, random_state = 42)

train_df = df[df['Combined'].isin(train_indices)].iloc[:,6:].drop(['Combined'], axis=1)
test_df = df[df['Combined'].isin(test_indices)].iloc[:,6:].drop(['Combined'], axis=1)

print(train_df.shape, test_df.shape)
print(test_indices[:5])
```

```python
train_df.columns
```

```python
cols = ['q75', 'q65', 'q55']
# cols = []
test_df = test_df.drop(cols, axis=1)
train_df = train_df.drop(cols, axis=1)
train_df
```

```python
from pycaret.classification import *
stp = setup(train_df, target = 'Is_pass', test_data=test_df)
```

```python
from catboost import CatBoostClassifier

X = stp.X_test
y = stp.y_test

X_train = stp.X_train
y_train = stp.y_train

# CatBoost Model
cat_model = CatBoostClassifier(eval_metric='Logloss', early_stopping_rounds=100, verbose = False)
cat_model.fit(X_train, y_train, eval_set=(X, y))
```

```python
from sklearn.metrics import f1_score, roc_auc_score, recall_score, precision_score, precision_recall_curve, accuracy_score

def metrics(model, X_test, y_test):
    X_pred_prob = model.predict_proba(X_test)[:, 1]
    X_pred = model.predict(X_test)
    
    f1 = f1_score(y_test, X_pred)
    auc = roc_auc_score(y_test, X_pred_prob)
    prec = precision_score(y_test, X_pred)
    rec = recall_score(y_test, X_pred)
    acc = accuracy_score(y_test, X_pred)
    
    print(f"Accuracy: {acc}")
    print(f"AUC: {auc}")
    print(f"Precision: {prec}")
    print(f"Recall: {rec}")
    print(f"F1: {f1}")

metrics(cat_model, X, y)
```

```python
def best_f1(model, X_test, y_test):
    X_pred_prob = model.predict_proba(X_test)[:, 1]
    # Вычисляем precision, recall и thresholds
    precision, recall, thresholds = precision_recall_curve(y_test, X_pred_prob)
    
    # Вычисляем F1-score для каждого порога
    f1_scores = [2 * (p * r) / (p + r) for p, r in zip(precision, recall)]
    
    # Находим порог, который максимизирует F1-score
    best_id = np.argmax(f1_scores)
    best_threshold = thresholds[best_id]
    best_f1 = np.max(f1_scores)
    best_prec = precision[best_id]
    best_rec = recall[best_id]
    
    print(f"Лучший порог: {best_threshold}")
    print(f"Лучшая точность (precision): {best_prec}")
    print(f"Лучшая полнота (recall): {best_rec}")
    print(f"Лучший F1-score: {best_f1}")

best_f1(cat_model, X, y)
```

```python
def pipeline(s):
    return s.iloc[:,6:].drop(['Combined', 'Is_pass']+cols, axis=1)
```

```python
from tqdm import tqdm

accuracy_dict = {}

for name in tqdm(unique_values, desc="Processing programs"):
    program_df = df[df['Combined'].str.contains(name, regex=False)]
    program_X = pipeline(program_df)
    program_y = program_df["Is_pass"]

    program_pred = voting_clf.predict(program_X)
    program_accuracy = accuracy_score(program_y, program_pred)
    accuracy_dict[name] = program_accuracy

acc_df = pd.DataFrame({"Program": list(accuracy_dict.keys()), "Accuracy": list(accuracy_dict.values())})

acc_df["Is_test"] = acc_df["Program"].isin(test_indices)
```

```python
import matplotlib.pyplot as plt
def plot(name, a, s, is_test, i):
    n= 110
    m= 110
    
    pd_df = pd.DataFrame({col: [val]*n*m for col, val in a.items()})
    
    # Create the meshgrid
    c, d = np.meshgrid(np.linspace(0, 2.5, n), np.linspace(50,110, m))
    
    # Convert the meshgrid to a DataFrame
    pd_df['Orig_above'] = c.flatten()
    pd_df['Avg_score'] = d.flatten()
    
    y_pred_proba = voting_clf.predict_proba(pd_df)[:,1]
    
    
    quota = a['Quota']
    num = a['Num_subjects']
    # Plot the graph
    plt.figure(figsize=(5, 4))
    plt.contourf(c*quota, d*num, y_pred_proba.reshape(m, n), cmap='RdBu_r')
    plt.scatter(x= quota*s['Orig_above'],y=num*s['Avg_score'], c= s['Is_pass'], cmap='seismic')
    plt.colorbar(label='Prediction')
    plt.xlabel('Orig_above')
    plt.ylabel('Score')
    plt.axvline(x=a['Quota'], color='k', linestyle='--', label='Quota')
    plt.title(name)
    
    path = "Assets/"
    if is_test: path+="Test/"
    else: path+="Train/"
    path+=f"{name[:120]}.png".replace("..",".")
    plt.savefig(path, dpi=300, bbox_inches='tight')
    if i%500==0: plt.show()
    plt.close()
```

```python
for i, name in tqdm(enumerate(unique_values), total=len(unique_values)):
    is_test = False
    if name in test_indices: is_test = True
    s = df[df['Combined'].str.contains(name, regex=False)]
    s = s[s['Avg_score'] > 50]
    s = s[s['Orig_above']< 2.5]
    a = s.iloc[0,6:].drop(['Combined', 'Is_pass']+cols)
    plot(name, a, s, is_test, i)
```
